```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
import ast 
def clean_string(data):
    if pd.notna(data): 
        return ast.literal_eval(data) 
```

```python
df['job_skills']=df['job_skills'].apply(clean_string)
```

```python
df_salary=df[['job_title_short','salary_year_avg']].copy()
def inflation(data): 
    if 'Senior' in data['job_title_short']: 
       return data['salary_year_avg']*1.05 
    else:
         return data['salary_year_avg']*1.03 
```

```python
df_salary['salary_year_avg_inflated']=df_salary.apply(inflation,axis=1)
```
