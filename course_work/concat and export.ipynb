```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df['job_posted_month']=df['job_posted_date'].dt.month_name()
```

```python
#month=["January","Febuary","March","April","May","June","July","August","September","October","Novermber","December"]\
month=df['job_posted_month'].unique()
month
```

```python
job_month={i:df[df['job_posted_month']==i] for i in month}

temp=pd.concat([job_month["January"],job_month["February"],job_month["March"]],ignore_index=True)
```

```python
temp['job_posted_month'].value_counts().plot(kind='bar')
```

```python
temp.to_excel('quater1_data.xlsx')
```
