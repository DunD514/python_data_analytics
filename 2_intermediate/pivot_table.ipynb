```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python

hold=df['job_country'].value_counts().sort_values(ascending=False).head(6).index

df_job_country=df.pivot_table(
    values='salary_year_avg',
    index='job_country',
    columns='job_title_short',
    aggfunc='median'
)
```

```python
df_job_country=df_job_country.loc[hold]
hold
```

```python
titles=['Data Analyst','Data Scientist','Data Engineer']
```

```python
df_job_country=df_job_country[titles]
```

```python
df_job_country.plot(kind='bar')
```
