```python
# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df['job_work_from_home'].value_counts().plot(kind='pie',startangle=90,autopct='%1.2f%%')
```

```python
dictionary={'job_work_from_home':'job__no_work_from_home',
            'job_no_degree_mention':'job_degree mention',
            'job_health_insurance':'job_no_health_insurance'
            }
fig,ax=plt.subplots(1,3)
for i,(key,val) in enumerate(dictionary.items()): 
    df[key].value_counts().plot(kind='pie',ax=ax[i],autopct='%1.2f%%',startangle=90)
    ax[i].set_ylabel("")
```
